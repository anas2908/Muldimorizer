```python
import os
import pandas as pd

# Input folder containing .tsv files
input_folder = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\Raw-Text"
# Output folder where .xlsx files will be saved
output_folder = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\preprocssed_excel"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop through each file in the input folder
for file_name in os.listdir(input_folder):
    if file_name.endswith('.tsv'):
        # Construct full file path
        tsv_file_path = os.path.join(input_folder, file_name)
        
        # Load the .tsv file into a DataFrame
        df = pd.read_csv(tsv_file_path, sep='\t')
        
        # Create a new column 'file_name' with the name of the file (without extension)
        df['file_name'] = os.path.splitext(file_name)[0]
        
        # Construct output .xlsx file path
        xlsx_file_name = os.path.splitext(file_name)[0] + '.xlsx'
        xlsx_file_path = os.path.join(output_folder, xlsx_file_name)
        
        # Save the DataFrame as an Excel file
        df.to_excel(xlsx_file_path, index=False)

print("All .tsv files have been converted and saved as .xlsx with the file name column.")
```

now write a code to take an excel sheet and a folder inside the folder go to sub folder whose name matches name of 'episode' column of excel, then within the sub folder got to the sub-sub-folder whose name matches the name of 'scene' column then within the sub sub folder chose the file whose name matches with the excel column 'part'.wav  convert this .wav file to mffcs embeddings and save it in a new folder with the name as 'uid' in the excel sheet

```python
import os
import pandas as pd
import librosa
import numpy as np

# Function to extract MFCC embeddings
def extract_mfcc(file_path, n_mfcc=13):
    # Load the .wav file
    y, sr = librosa.load(file_path, sr=None)
    
    # Extract MFCC features
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    
    # Return the MFCC features as a numpy array
    return mfccs

# Input Excel sheet path
excel_file = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\preprocssed_excel\MasterExcel.xlsx"
# Root folder containing episode/scene/part folders
root_folder = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\Raw-Audio"
# Output folder to save MFCC embeddings
output_folder = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\AudioPreprocessed"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Load the Excel sheet into a DataFrame
df = pd.read_excel(excel_file)

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    # Extract the necessary information from the Excel sheet
    episode = str(row['file_name'])
    scene = str(row['Scenes'])
    part = str(row['Sl. No.'])
    uid = str(row['uid'])
    
    # Construct the path to the target .wav file
    wav_file_path = os.path.join(root_folder, episode, scene, f"{part}.wav")
    
    # Check if the .wav file exists
    if os.path.exists(wav_file_path):
        # Extract MFCC embeddings
        mfcc_embeddings = extract_mfcc(wav_file_path)
        
        # Define the output path for the embeddings (saved as .npy file)
        output_file_path = os.path.join(output_folder, f"{uid}.npy")
        
        # Save the MFCC embeddings as a numpy array
        np.save(output_file_path, mfcc_embeddings)
        
        print(f"Processed and saved MFCC embeddings for UID: {uid}")
    else:
        print(f"File not found: {wav_file_path}")

print("All files processed.")
```

do the same for videos not audio file by that i mean, instead of .wav search for .mp4 file and then split the videos into 1fps  then apply clip for embeddings so by that i mean if i have 10sec video there will be 10 frames and since one frame has embeddings [1,512]  10 frames (whole video) will have [10,512] shape

```python
import os
import cv2
import torch
import numpy as np
from PIL import Image
from transformers import CLIPProcessor, CLIPModel
import pandas as pd

# Load CLIP model and processor
device = "cuda" if torch.cuda.is_available() else "cpu"
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Function to extract frames from a video at 1 fps
def extract_frames(video_path, fps=1):
    vidcap = cv2.VideoCapture(video_path)
    success, image = vidcap.read()
    frame_list = []
    count = 0
    while success:
        if count % int(vidcap.get(cv2.CAP_PROP_FPS)) == 0:  # Capture one frame per second
            frame_list.append(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert to RGB for CLIP
        success, image = vidcap.read()
        count += 1
    vidcap.release()
    return frame_list

# Function to extract CLIP embeddings from frames
def extract_clip_embeddings(frames):
    embeddings_list = []
    for frame in frames:
        pil_image = Image.fromarray(frame)
        inputs = processor(images=pil_image, return_tensors="pt").to(device)
        with torch.no_grad():
            embeddings = model.get_image_features(**inputs)
        embeddings_list.append(embeddings.cpu().numpy().flatten())  # Flatten to [1, 512]
    return np.array(embeddings_list)  # Shape [n_frames, 512]

# Input Excel sheet path
excel_file = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\preprocssed_excel\MasterExcel.xlsx"
# Root folder containing episode/scene/part folders
root_folder = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\Raw-Visual"
# Output folder to save embeddings
output_folder = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\VideoPreprocessed"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Load the Excel sheet into a DataFrame
df = pd.read_excel(excel_file)

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    # Extract the necessary information from the Excel sheet
    episode = str(row['file_name'])
    scene = str(row['Scenes'])
    part = str(row['Sl. No.'])
    uid = str(row['uid'])
    
    # Construct the path to the target .mp4 file
    video_file_path = os.path.join(root_folder, episode, scene, f"{part}.mp4")
    
    # Check if the .mp4 file exists
    if os.path.exists(video_file_path):
        # Step 1: Extract frames from the video at 1 fps
        frames = extract_frames(video_file_path, fps=1)
        
        # Step 2: Extract CLIP embeddings from the frames
        clip_embeddings = extract_clip_embeddings(frames)
        
        # Step 3: Define the output path for the embeddings (saved as .npy file)
        output_file_path = os.path.join(output_folder, f"{uid}.npy")
        
        # Step 4: Save the embeddings as a numpy array
        np.save(output_file_path, clip_embeddings)
        
        print(f"Processed and saved CLIP embeddings for UID: {uid}, video: {video_file_path}")
    else:
        print(f"File not found: {video_file_path}")

print("All files processed.")
```

```python
import os
import cv2
import pandas as pd
import base64
from openai import OpenAI

# Initialize the OpenAI client with your API key
# client = OpenAI(api_key='')
client = OpenAI(api_key='')
# Function to encode the image
# def encode_image(image_path):
#     with open(image_path, "rb") as image_file:
#         return base64.b64encode(image_file.read()).decode('utf-8')
def encode_image(image_path):
    try:
        with open(image_path, "rb") as image_file:
            return base64.b64encode(image_file.read()).decode('utf-8')
    except FileNotFoundError:
        print(f"File not found: {image_path}")
        return None
    except Exception as e:
        print(f"Error encoding image: {e}")
        return None

# Function to extract the middle frame from the video
def extract_middle_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    middle_frame = total_frames // 2
    cap.set(cv2.CAP_PROP_POS_FRAMES, middle_frame)
    success, frame = cap.read()
    if success:
        frame_path = video_path.replace('.mp4', '_middle_frame.jpg')
        cv2.imwrite(frame_path, frame)
        cap.release()
        return frame_path
    cap.release()
    return None

# Function to summarize the image
def summarize_image(base64_image, speaker, utterance, goldlabel):
    humor_label = "humorous" if goldlabel == "humor" else "non-humorous"
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "user",
                "content": f"The speaker {speaker} says: '{utterance}'. "
                           f"Given this image, explain why the scene is {humor_label}, in only one short paragraph, also even if you are confused regarding the explanation give some generic explanation. In the summary do include the the speaker information and what he says also how his facial expression is. Very Imprtantly keep it a very simple english"
            },
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": "Please analyze this frame.",
                    },
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/jpeg;base64,{base64_image}"
                        }
                    }
                ]
            }
        ]
    )
    return response.choices[0].message.content

# Function to process videos and summarize based on the Excel sheet
def process_videos_and_summarize(excel_path, root_folder):
    df = pd.read_excel(excel_path)
    
    for index, row in df.iterrows():
        # Check if summary already exists
        if 'summary' in df.columns and pd.notna(row['summary']):
            print(f"Summary for {row['uid']} already exists. Skipping...")
            continue
        
        # Extract the necessary information from the Excel sheet
        episode = str(row['file_name'])
        scene = str(row['scene'])
        part = str(row['part'])
        speaker = str(row['speaker'])  # Assuming the column names are as specified
        utterance = str(row['english'])
        goldlabel = str(row['goldlabel'])
        uid = str(row['uid'])

        # Construct the path to the target .mp4 file
        video_file_path = os.path.join(root_folder, episode, scene, f"{part}.mp4")
        
        frame_path = extract_middle_frame(video_file_path)
        
        if frame_path:
            base64_image = encode_image(frame_path)
            summary = summarize_image(base64_image, speaker, utterance, goldlabel)
            print(f"{summary}")  # Print the summary
            
            # Save the summary immediately to the DataFrame and Excel file
            df.at[index, 'summary'] = summary
            df.to_excel(excel_path, index=False)  # Save after each summary
        else:
            df.at[index, 'summary'] = "Frame extraction failed."
            df.to_excel(excel_path, index=False)  # Save failure notice immediately

# Example usage
excel_path = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\preprocssed_excel\MasterExcel.xlsx"
root_folder = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\Raw-Visual"
process_videos_and_summarize(excel_path, root_folder)
```
