```python
import os
import cv2
import pandas as pd
import base64
from openai import OpenAI

# Initialize the OpenAI client with your API key
# client = OpenAI(api_key='')
client = OpenAI(api_key='')
# Function to encode the image
# def encode_image(image_path):
#     with open(image_path, "rb") as image_file:
#         return base64.b64encode(image_file.read()).decode('utf-8')
def encode_image(image_path):
    try:
        with open(image_path, "rb") as image_file:
            return base64.b64encode(image_file.read()).decode('utf-8')
    except FileNotFoundError:
        print(f"File not found: {image_path}")
        return None
    except Exception as e:
        print(f"Error encoding image: {e}")
        return None

# Function to extract the middle frame from the video
def extract_middle_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    middle_frame = total_frames // 2
    cap.set(cv2.CAP_PROP_POS_FRAMES, middle_frame)
    success, frame = cap.read()
    if success:
        frame_path = video_path.replace('.mp4', '_middle_frame.jpg')
        cv2.imwrite(frame_path, frame)
        cap.release()
        return frame_path
    cap.release()
    return None

# Function to summarize the image
def summarize_image(base64_image, speaker, utterance, goldlabel):
    humor_label = "humorous" if goldlabel == "humor" else "non-humorous"
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "user",
                "content": f"The speaker {speaker} says: '{utterance}'. "
                           f"Given this image, explain in only one short paragraph if its humoros and why."
            },
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": "Please analyze this frame.",
                    },
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/jpeg;base64,{base64_image}"
                        }
                    }
                ]
            }
        ]
    )
    return response.choices[0].message.content

# Function to process videos and summarize based on the Excel sheet
def process_videos_and_summarize(excel_path, root_folder):
    df = pd.read_excel(excel_path)
    
    for index, row in df.iterrows():
        # Check if summary already exists
        if 'summary_chatgpt' in df.columns and pd.notna(row['summary_chatgpt']):
            print(f"Summary for {row['uid']} already exists. Skipping...")
            continue
        
        # Extract the necessary information from the Excel sheet
        episode = str(row['file_name'])
        scene = str(row['scene'])
        part = str(row['part'])
        speaker = str(row['speaker'])  # Assuming the column names are as specified
        utterance = str(row['english'])
        goldlabel = str(row['goldlabel'])
        uid = str(row['uid'])

        # Construct the path to the target .mp4 file
        video_file_path = os.path.join(root_folder, episode, scene, f"{part}.mp4")
        
        frame_path = extract_middle_frame(video_file_path)
        
        if frame_path:
            base64_image = encode_image(frame_path)
            summary = summarize_image(base64_image, speaker, utterance, goldlabel)
            print(f"{summary}")  # Print the summary
            
            # Save the summary immediately to the DataFrame and Excel file
            df.at[index, 'summary_chatgpt'] = summary
            df.to_excel(excel_path, index=False)  # Save after each summary
        else:
            df.at[index, 'summary_chatgpt'] = "Frame extraction failed."
            df.to_excel(excel_path, index=False)  # Save failure notice immediately

# Example usage
excel_path = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\preprocssed_excel\MasterExcel_test.xlsx"
root_folder = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\Raw-Visual"
process_videos_and_summarize(excel_path, root_folder)
```

```python
import pandas as pd
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

# Load the Excel file
file_path = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\preprocssed_excel\MasterExcel_test.xlsx"
df = pd.read_excel(file_path)

# Extract the 'summary' (true labels) and 'summary_chatgpt' (predicted labels) columns
y_true = df['summary'].astype(str)
y_pred = df['summary_chatgpt'].astype(str)

# Initialize ROUGE scorer (using ROUGE-1, ROUGE-2, and ROUGE-L)
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# Initialize BLEU scoring function
smoothing_function = SmoothingFunction().method4

# Store ROUGE and BLEU scores
rouge1_scores = []
rouge2_scores = []
rougeL_scores = []
bleu1_scores = []
bleu2_scores = []
bleu3_scores = []
bleu4_scores = []

# Calculate ROUGE and BLEU scores for each pair of true and predicted summaries
for true, pred in zip(y_true, y_pred):
    # ROUGE Scores
    rouge1 = scorer.score(true, pred)['rouge1'].fmeasure
    rouge2 = scorer.score(true, pred)['rouge2'].fmeasure
    rougeL = scorer.score(true, pred)['rougeL'].fmeasure
    
    rouge1_scores.append(rouge1)
    rouge2_scores.append(rouge2)
    rougeL_scores.append(rougeL)
    
    # BLEU Scores (tokenize the sentences and calculate BLEU-1, BLEU-2, BLEU-3, BLEU-4)
    reference = true.split()  # Tokenize the reference summary
    hypothesis = pred.split()  # Tokenize the predicted summary
    
    bleu1 = sentence_bleu([reference], hypothesis, weights=(1, 0, 0, 0), smoothing_function=smoothing_function)
    bleu2 = sentence_bleu([reference], hypothesis, weights=(0.5, 0.5, 0, 0), smoothing_function=smoothing_function)
    bleu3 = sentence_bleu([reference], hypothesis, weights=(0.33, 0.33, 0.33, 0), smoothing_function=smoothing_function)
    bleu4 = sentence_bleu([reference], hypothesis, weights=(0.25, 0.25, 0.25, 0.25), smoothing_function=smoothing_function)
    
    bleu1_scores.append(bleu1)
    bleu2_scores.append(bleu2)
    bleu3_scores.append(bleu3)
    bleu4_scores.append(bleu4)

# Calculate the average of each score
average_rouge1 = sum(rouge1_scores) / len(rouge1_scores)
average_rouge2 = sum(rouge2_scores) / len(rouge2_scores)
average_rougeL = sum(rougeL_scores) / len(rougeL_scores)

average_bleu1 = sum(bleu1_scores) / len(bleu1_scores)
average_bleu2 = sum(bleu2_scores) / len(bleu2_scores)
average_bleu3 = sum(bleu3_scores) / len(bleu3_scores)
average_bleu4 = sum(bleu4_scores) / len(bleu4_scores)

# Print the average scores
print(f"Average ROUGE-1 F1-score: {average_rouge1:.4f}")
print(f"Average ROUGE-2 F1-score: {average_rouge2:.4f}")
print(f"Average ROUGE-L F1-score: {average_rougeL:.4f}")

print(f"Average BLEU-1 score: {average_bleu1:.4f}")
print(f"Average BLEU-2 score: {average_bleu2:.4f}")
print(f"Average BLEU-3 score: {average_bleu3:.4f}")
print(f"Average BLEU-4 score: {average_bleu4:.4f}")
```

```python
import pandas as pd
import json

# Function to convert JSON to DataFrame and save it to Excel
def json_to_excel(json_file, excel_file):
    # Load the JSON data
    with open(json_file, 'r') as f:
        data = json.load(f)
    
    # Initialize an empty list to hold the extracted data
    extracted_data = []

    # Iterate over the dialogs in the JSON
    for dialog_key, dialog_info in data.items():
        scene = dialog_info.get('Scene', None)
        gt = dialog_info.get('Ground Truth', None)  # Assuming 'GT' field is present for ground truth

        # Iterate through the dialog turns
        for dialog_turn_key, dialog_turn_info in dialog_info.items():
            if dialog_turn_key.startswith('Dialog Turn'):
                speaker = dialog_turn_info.get('Speaker', None)
                dialogue = dialog_turn_info.get('Dialog', None)

                # Append the data to the list, including GT
                extracted_data.append({
                    'Scene': scene,
                    'Speaker': speaker,
                    'Dialogue': dialogue,
                    'Ground Truth': gt  # Including the ground truth (GT)
                })
    
    # Convert the list of dictionaries into a DataFrame
    df = pd.DataFrame(extracted_data)

    # Save the DataFrame to an Excel file
    df.to_excel(excel_file, index=False)
    print(f"Excel file saved at: {excel_file}")

# Example usage
json_file = r"C:\Users\Anas Khan\Desktop\example_dataset.json"  # Path to your input JSON file
excel_file = r"C:\Users\Anas Khan\Desktop\example_dataset.xlsx"  # Path where you want to save the Excel file

json_to_excel(json_file, excel_file)
```

```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score
import random

# Load the Excel file
excel_path = r"C:\Users\Anas Khan\Desktop\shemud\M2H2-dataset-main\Main-Dataset\preprocssed_excel\TBBT_dataset.xlsx"
df = pd.read_excel(excel_path)

# Define the column for ground truth values
actual_column = 'Ground Truth'  # Ground truth values

# Check if the ground truth column contains binary values (0 or 1)
if df[actual_column].isin([0, 1]).all():
    num_iterations = 10  # Define the number of random iterations
    results = []

    for i in range(num_iterations):
        # Generate a random prediction (0 or 1) for each row in the dataset
        random_predictions = [random.choice([0, 1]) for _ in range(len(df))]
        
        # Calculate Precision, Recall, and F1 Score
        precision = precision_score(df[actual_column], random_predictions)
        recall = recall_score(df[actual_column], random_predictions)
        f1 = f1_score(df[actual_column], random_predictions)
        
        # Store the results for each iteration
        results.append({
            'Iteration': i + 1,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1
        })
        
        print(f"Iteration {i + 1} - Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

    # Optionally, convert results to a DataFrame for easier viewing
    results_df = pd.DataFrame(results)
    print("\nAverage Scores Over Iterations:")
    print(results_df.mean(numeric_only=True))
else:
    print("The 'Ground Truth' column must contain binary values (0 or 1).")
```
